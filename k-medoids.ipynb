```python
import random
import matplotlib.pyplot as plt
```

```python
points = [[(2,8), (2,5), (8,4), (5,8), (7,5), (6,4), (1,2), (4,3)]]
k=2
```

```python
random.shuffle(points)
```

```python
def map_plot(cluster,k):
    for i in range(k):
        x_val= [x[0] for x in cluster[i]]
        y_val = [x[1] for x in cluster[i]]
        plt.scatter(x_val, y_val)
```

```python
map_plot(points,1)
plt.title("Initial points")
plt.show()
```

```python
def calculate_cost(a_medoid,b_medoid):
    cost=0
    clusters=[[a_medoid],[b_medoid]]
    for x in points[0]:
        if(x==a_medoid or x==b_medoid):
            continue
        a_dist=abs(x[0]-a_medoid[0])+abs(x[1]-a_medoid[1])
        b_dist=abs(x[0]-b_medoid[0])+abs(x[1]-b_medoid[1])
        if(a_dist<b_dist):
            cost+=a_dist
            clusters[0].append(x)
        else:
            cost+=b_dist
            clusters[1].append(x)
    # print(cost, clusters)
    return (cost,clusters)
```

```python
min_cost=999999
optimal_clusters=[]
for idx,a_medoid in enumerate(points[0]):
    # print(idx, a_medoid)
    for b_medoid in points[0][idx+1:]:
        cost,cluster = calculate_cost(a_medoid, b_medoid)
        if(cost<min_cost):
            min_cost=cost
            optimal_clusters=cluster
print(optimal_clusters)
map_plot(optimal_clusters,k)
plt.title("optimal clusters")
plt.show()

# clusters
```
